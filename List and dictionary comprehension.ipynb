```python
a={'kamal':20,'teja':30,'sai':60}
b={i:j*2 for (i,j) in a.items() if j>=30}
print(b)
```

```python
a={'sai':60,'kamal':40,'teja':90}
b={}
for i,j in a.items():
    if j>=50:
        b[i]='pass'
    else:
        b[i]='fail'
print(b)        
```
