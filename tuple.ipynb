```python
even=[]
odd=[]
for i in range(1,12):
    if i%2==0:
        even.append(i++2)
    else:
        odd.append(i++3)
print('even values=',even)
print('odd values=',odd)
```

```python
a={'teja':23,'vissu':34}
for i in a.keys():
    print('keys of the dictionary=',i)
for j in a.values():
    print('values of the dictionary',i)
for i,j in a.items():
    print('keys',i)
    print('values',j)
    
```
